```python
from operator import concat
import twint

c = twint.Config()
c.Search = "work OR labor OR job OR employment OR occupation OR business"
c.Limit = 10000
c.Store_csv = True
c.Output = concat(concat("tweets_",c.Search),".csv")
c.Since = "2020-02-23 00:00:00"
c.Lang = 'en'
c.Debug = True

try:
    twint.run.Search(c)
    #tweets_list = twint.output.tweets_list[:20]
    #print(tweets_list)
except Exception as e:
    print('Error:',e)
```
