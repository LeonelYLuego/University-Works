```python
from IPython.core.interactiveshell import SeparateUnicode
import numpy as ny
import random
from timeit import default_timer
inicio=default_timer()
nuevaP=[]
Nproc=8
def Conbinaciones(MTWS,TWT):
    combinacion=[]
    
    #basicamente agregamos valore hasta que ya no quepan
    while (sum(combinacion)+MTWS)< TWT:
        combinacion.append(random.randint(1,MTWS))
    combinacion.append(TWT-sum(combinacion))
    return combinacion

nuevaP=Conbinaciones(200,1000)
proc=[]
Lproc=[]
tempo=[]
pob=[]
copiaP=nuevaP;
for i in range(20):

    for j in range(Nproc-1):#nos da los numeros para partir elvector en la cantidad de procesadores
      proc.append(random.randint(1,len(nuevaP)-1))
    proc.sort()#los ordena
    Lproc.append(nuevaP[:proc[0]])
    for k in range(0,len(proc)-1):#crea vectores de lo quecontiene cada procesador
      Lproc.append(nuevaP[proc[k]:proc[k+1]])
    Lproc.append(nuevaP[proc[-1]:len(nuevaP)])
    pob.append(Lproc[:])#agregamos al individuo a la poblacion
    random.shuffle(nuevaP)
    Lproc.clear()#limpiamos
    proc.clear()
lmay=[]
for i in pob:
  may=0
  #print(i)#imprimimos individuo
  for n in i:
    if sum(n)>may:
      may=sum(n)
  lmay.append(may)

print("Padres:")
Antpadres = []

for i in range(2):
  Antpadres.append(pob[lmay.index(min(lmay))])
  print(Antpadres[i])
  lmay.remove(min(lmay))

padres = []
 
for pad in Antpadres:
  temp=[]
  for i in pad:
    for j in i:
      temp.append(j)
  padres.append(temp)
print(padres)

flagsPadres = ny.full((len(padres[0])), False)
print(flagsPadres)

start = random.randint(0,len(padres[0])-1)
tempStart = start
print("START: ",start)
while padres[0][start] != padres[1][tempStart]:
  flagsPadres[tempStart] = True
  i = 0
  for e in padres[0]:
    if e == padres[1][tempStart]:
      tempStart = i
      print(tempStart)
      break
    i = i + 1










```
